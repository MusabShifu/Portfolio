```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load and describe data

```python
sales = pd.read_excel("superstore_sales.xlsx")
```

```python
sales.describe()
```

```python
sales.info()
```

### WHAT IS THE OVERALL SALES TREND?

```python
to_ym = lambda x: str(x)[:7]
sales['year_month'] = sales['order_date'].apply(to_ym)
sales_trend = sales.groupby('year_month')['sales'].sum()
```

```python
plt.figure(figsize=(14,6))
plt.plot(sales_trend.index , sales_trend.values , color='green')
plt.xticks(rotation='vertical' , size=10)
plt.title("Sales Trend" , size=22)
plt.show()
```

### WHICH ARE THE TOP 10 PRODUCTS BY SALES

```python
top_10_product = sales.groupby('product_name').sum()['sales'].sort_values(ascending=0)[:10]
top_10_product
```

```python
plt.figure(figsize=(14,8))
sns.barplot(y = top_10_product.index , x = top_10_product.values)
plt.xticks(rotation=45, size=10)
plt.title("Top 10  Products By Sales" , size=22)
plt.xlabel('Sales', size=18)
plt.ylabel('Products', size=18)
plt.show()
```

### WHICH ARE THE MOST SELLING PRODUCTS

```python
most_selling = sales.groupby("product_name")['quantity'].sum().sort_values(ascending = 0)[:10]
most_selling 
```

```python
plt.figure(figsize=(14,8))
sns.barplot(y = most_selling.index , x = most_selling.values)
plt.xticks(rotation=45, size=10)
plt.title("Top 10 Saled Products" , size=22)
plt.xlabel('Sellings', size=18)
plt.ylabel('Products', size=18)
plt.show()
```

### WHAT IS THE MOST PREFERRED SHIP MODE

```python
sales_count = sales.copy()
sales_count['count'] = 1
most_ship_mode = sales_count.groupby("ship_mode")['count'].sum()
most_ship_mode
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x='ship_mode' , data=sales)
plt.title("Ship Modes Count", size=22)
plt.xlabel('Ship Mode', size=10)
plt.show()
```

### WHAT IS AVERAGE PROFIT FOR EACH YEAR

```python
average_profit_per_year = sales.groupby('year')['profit'].mean()
average_profit_per_year
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x=average_profit_per_year.index , y= average_profit_per_year.values)
plt.title("Average Profit Per Year", size=22)
plt.xlabel('Year', size=16)
plt.ylabel('Profit', size=16)
plt.show()
```
